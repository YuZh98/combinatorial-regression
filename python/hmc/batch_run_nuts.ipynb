```python
from nuts_ilp_model import MCMCConfig, DiagnosticConfig, run_experiment
import numpy as np
```

```python
list_n = [1000]
list_d = [5, 10]
list_p = [5]
list_m = [1, 2, 5, 10, 20, 50, 100]

out_dir = "../../results/runs/hmc"
mcmc_config = MCMCConfig(num_warmup=1000, num_samples=2000, num_chains=1, target_accept_prob=0.75)
diagnostic_config = DiagnosticConfig(max_plots=8, max_lag=50)
alg = "marginal_joint_green"
for n in list_n:
    for d in list_d:
        for p in list_p:
            for m in list_m:
                if m < d:
                    try:
                        mcmc, samples, diagnostics = run_experiment(
                            algorithm=alg,
                            n=n, p=p, d=d, m=m,
                            tau_beta=1.0,
                            mcmc_config=mcmc_config,
                            diagnostic_config=diagnostic_config,
                            data_seed=123,
                            mcmc_seed=456,
                            use_simulated_data=True,
                            save_results=True,
                            run_diagnostics=True,
                            output_dir=out_dir,
                        )
                        print(f"\n✓ {alg.upper()} completed successfully\n")
                    except Exception as e:
                        print(f"\n✗ {alg.upper()} failed: {e}\n")
                        import traceback
                        traceback.print_exc()

                    
                    
                    
                    
```
